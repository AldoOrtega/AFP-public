## Generando la columna RUT a partir de RUTDV

Como se vió en el notebook 2. Visualización de clientes, la tabla CCV no posee la columna RUT. RUT es el identificador individual de un ciudadano chileno a nivel nacional, normalmente esta compuesto por un número seguido de un guión y carácter que puede ser un dígito o una letra k, llamado dígito verificador. Este dígito se utiliza para comprobar la veracidad del RUT.

Entonces, la columna RUT es  igual a la columna RUTDV, sin el último carácter (dígito verificador). En esta representación del RUT el número y el dígito verificador se encuentran juntos, sin ser separados por un guión.

```python
ccv['RUT'] = ccv['RUTDV'].str[:-1]
# ccv.head()
ccv['RUT'] = ccv['RUT'].astype('int')

ccv.dtypes
```

```python
productos = productos.append(ccv, sort=False)
productos.drop(columns=cols_productos).head()
```

```python
# Cargando clientes, solo aquellas columnas que aportan información

clientes = pd.DataFrame()

for i in [1,2,3]:
    file = '../AFP data/Clientes/cl'+str(i)+'v3.csv'
    print("cargando clientes "+str(i))
    cl = pd.read_csv(file, encoding='latin-1')
    clientes = clientes.append(cl[['RUT','TIPO_SEXO','EDAD','PAIS_NACIMIENTO',
                                   'PAIS_NACIONALIDAD','PROFESION','AFP_ACTUAL','TIPO_TRABAJADOR',
                                   'RENTA_IMPONIBLE','COMUNA','VIGENTE']])
    
del cl
clientes.drop(columns=cols_clientes).head()
```

Como el RUT de los clientes en los productos esta sin dígito verificardor, es necesario sacarle el digito al RUT al DF clientes para poder hacer el matching. En esta ocación, el número del RUT y su dígito verificador se encuentran separados por un guión (-).

```python
clientes['RUT'] = clientes['RUT'].str.split('-', 1).str[0]
#clientes.head()
```

```python
clientes['RUT'] = clientes['RUT'].astype('int')
```

```python
clientes.dtypes
```

## Primer enfoque de estructura de datos

En una primera instancia, se unirán clientes y productos solo con variables binarias que indiquen si un cliente posee o no contratado cada uno de los cuatro productos

```python
# Creando una variable binaria asociada a cada producto indiccando
# si este producto se encuentra actualmente contratado o no
 
productos['ACTUAL'] =  ((productos['FONDO_A']>0) | 
                        (productos['FONDO_B']>0) | 
                        (productos['FONDO_C']>0) | 
                        (productos['FONDO_D']>0))
productos.drop(columns=cols_productos).head()
```

```python
import numpy as np

productos['BIN']  = np.select([productos.ACTUAL == True, productos.ACTUAL == False],[1,0])

productos.drop(columns=cols_productos).head()
```

## Uniendo tablas utilizando merge

En las primeras celdas se crea el df merge-actual.csv en donde se representa los productos que los clientes tienen contratados actualmente (posee saldo mayor que 0 en algunos de los fondos asociados a una cuenta), mientras que en la segunda parte se crea el archivo merge-ever.csv que representa los productos que los clientes han contratados alguna vez (independientemente si ya no lo tienen contratado), es decir, tiene una cuenta creada pero con saldo 0 (tuvo saldo en algún momento pero ya fue retirado).

### merge-actual.csv

```python
import copy

df = copy.deepcopy(clientes)
```

```python
%%time

for cuenta in ['CCO', 'CCV', 'CDC', 'CAV']:

    df = df.merge(productos[['RUT','BIN']][productos.PRODUCTO == cuenta],
                       how='left')
    df['BIN'] = df['BIN'].fillna(0)
    df['BIN'] = df['BIN'].astype('int')
    df = df.rename(columns={'BIN': cuenta})
```

### merge-ever.csv

```python
productos['BIN_2'] = 1
```

```python
import copy

df = copy.deepcopy(clientes)

for cuenta in ['CCO', 'CCV', 'CDC', 'CAV']:

    df = df.merge(productos[['RUT','BIN_2']][productos.PRODUCTO == cuenta],
                       how='left')
    df['BIN_2'] = df['BIN_2'].fillna(0)
    df['BIN_2'] = df['BIN_2'].astype('int')
    df = df.rename(columns={'BIN_2': cuenta})
```
